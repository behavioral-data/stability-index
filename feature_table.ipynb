```python
def get_user_averages(behavior, behavior_name):
    user_mean_averages = []
#     user_std_averages = []
    for u in users:
        behavior_user = behavior.loc[behavior.uid == u]
        behavior_user = behavior_user.set_index('time')
        behavior_user = behavior_user[[behavior_name]]
        behavior_user = behavior_user.resample('D').sum()
        user_mean_averages.append(behavior_user[behavior_name].mean())
#         user_std_averages.append(behavior_user.std())
    return user_mean_averages
        
behaviors = [
             calllog,
             smslog,
             smslog,
             unlock, 
             conver, 
             sleep, 
             act,
             act,
             act,
             act,
             act,
             act,
             light, 
             audio_amp,
             audio_infer,
             audio_infer,
             audio_infer
            ]
behavior_names = [
                  'calllog',
                  'sms_in',
                  'sms_out',
                  'unlock',
                  'conver',
                  'sleep',
                  'act_invehicle',
                  'act_onbike',
                  'act_onfoot',
                  'act_still',
                  'act_tilting',
                  'act_unknown',
                  'light',
                  'audio_amp',
                  'audio_noise',
                  'audio_silence',
                  'audio_voice'                  
]

user_averages_table = pd.DataFrame()
for i in range(len(behaviors)):
    user_averages_table[behavior_names[i]] = get_user_averages(behaviors[i], behavior_names[i])
    print('finished behavior: ' + behavior_names[i])
    print(datetime.datetime.now()) 
    print()
user_averages_table.head()
user_averages_table.to_csv('feature_table_raw.csv')
```
